```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[ 1,  2],
              [ 1,  3],
              [ 1,  4],
              [ 1,  5],
              [ 1,  6]])
```

```python
X.shape
```

```python
y = np.array([75, 82, 85, 88 ,92])
```

```python
plt.scatter(X[:, 1], y)
```

```python
y_pred1 = 5 * X[:, 1] + 35 * X[:, 0] 
y_pred2 = 7.5 * X[:, 1] + 40 * X[:, 0]
```

```python
plt.scatter(X[:, 1], y)
plt.plot(X[:, 1], y_pred1, label='1')
plt.plot(X[:, 1], y_pred2, label='2')
plt.legend()
plt.show()
```
